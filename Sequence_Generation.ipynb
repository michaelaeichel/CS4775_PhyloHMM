## Using Seq Gen to Generate Synthetic Data to Use for PhyloHMM and Benchmarking

### Validate tree structure with Phylo

```python
from Bio import Phylo
from io import StringIO
```

```python
tree = Phylo.read("phylohmm_data_generation_tree.nwk", 'newick')
```

```python
Phylo.draw(tree)
```

### Generate synthetic nucleotide sequences corresponding to that tree

Sequence length 1000, random seed 4775

```python
%%sh
echo "(((Canis_lupus_familiaris:0.1,Canis_latrans:0.1):0.2,Vulpes_lagopus:0.3):0.4, ((Felis_catus:0.1,Felis_chaus:0.1):0.2,((Mus_musculus:0.15,Rattus_norvegicus:0.15):0.25, ((Ursus_arctos:0.2,Ursus_maritimus:0.2):0.3,Homo_sapiens:0.5):0.4):0.3):0.5);" > phylohmm_data_generation_tree.nwk
seq-gen -mHKY -t 2.0 -l10000 -n1 -z4775< phylohmm_data_generation_tree.nwk
```
